INBUILT FUNCTION SHOWIN GHOW LOGISTIC REGRESSION WORKS WITHOUT USING SCIKIT-learn LIBRARIES

```python
from math import exp
# Creating the logistic regression model

# Helper function to normalize data
def normalize(X):
    return X - X.mean()

# Method to make predictions
def predict(X, b0, b1):
    return np.array([1 / (1 + exp(-1*b0 + -1*b1*x)) for x in X])

# Method to train the model
def logistic_regression(X, Y):

    X = normalize(X)

    # Initializing variables
    b0 = 0
    b1 = 0
    L = 0.001
    epochs = 300

    for epoch in range(epochs):
        y_pred_test = predict(X, b0, b1)
        D_b0 = -2 * sum((Y - y_pred_test) * y_pred_test * (1 - y_pred_test))  # Derivative of loss wrt b0
        D_b1 = -2 * sum(X * (Y - y_pred_test) * y_pred_test * (1 - y_pred_test))  # Derivative of loss wrt b1
        # Update b0 and b1
        b0 = b0 - L * D_b0
        b1 = b1 - L * D_b1
    
    return b0, b1
```
